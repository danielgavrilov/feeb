````python
# Setup - imports and database connection
import asyncio
from sqlalchemy import text, func, select, distinct
from sqlalchemy.ext.asyncio import AsyncSession
import pandas as pd
import json

# Import app modules
import sys
sys.path.insert(0, '..')
from app.database import AsyncSessionLocal, engine
from app.models import (
    Ingredient, Allergen, Product, Recipe, RecipeIngredient,
    ProductIngredient, IngredientAllergen, MenuUpload
)

# Helper function to run async queries in notebook
async def run_query(query_str):
    async with AsyncSessionLocal() as session:
        result = await session.execute(text(query_str))
        return pd.DataFrame(result.fetchall(), columns=result.keys())

async def run_orm_query(stmt):
    async with AsyncSessionLocal() as session:
        result = await session.execute(stmt)
        return result.scalars().all()

# Notebook-friendly async execution
def query(q):
    return asyncio.run(run_query(q))

print("✓ Setup complete")
````

## 1. Open Food Facts (OFF) Data Pipeline

### Overview
The OFF import pipeline (`apps/api/data_pipeline/import_off.py`) downloads and imports:
- **~3.5M products** from the OpenFoodFacts database
- **Allergens** extracted from product tags
- **Ingredients** extracted from product tags

### Import Process

**Step 1:** Download `openfoodfacts-products.jsonl.gz` (~10GB file)

**Step 2:** Extract unique allergens
- Scans ALL products for `allergens_tags` (e.g., `"en:gluten"`, `"en:milk"`)
- Creates canonical `Allergen` records

**Step 3:** Extract unique ingredients  
- Scans ALL products for `ingredients_tags` (e.g., `"en:wheat-flour"`, `"en:sugar"`)
- Creates canonical `Ingredient` records with `source='off'`

**Step 4:** Import products with relationships
- Creates `Product` records
- Links products to ingredients via `ProductIngredient` (with rank)
- Links products to allergens via `ProductAllergen` (contains/may_contain)

### Database Tables Created
- `allergen` - Canonical allergen taxonomy (e.g., "Gluten", "Milk")
- `ingredient` - Canonical ingredient taxonomy (e.g., "Wheat Flour", "Sugar") with `source='off'`
- `product` - Product catalog (barcode, name, brand)
- `product_ingredient` - Many-to-many: products ↔ ingredients
- `product_allergen` - Many-to-many: products ↔ allergens
- `ingredient_allergen` - Many-to-many: ingredients ↔ allergens

### Key Code Locations
```

apps/api/app/dal.py                           # insert_ingredient(), insert_allergen(), etc.
```

## 2. LLM Recipe Analysis Pipeline

### Overview
The menu upload pipeline (`apps/api/app/services/menu_upload.py`) uses Google Gemini to:
1. **Extract dishes** from uploaded menus (PDF/Image/URL)
2. **Deduce ingredients** for each dish
3. **Identify allergens** for each ingredient

### Pipeline Stages

#### **Stage 0: Upload**
- User uploads a menu (PDF, image, or URL)
- File stored in `uploaded_menus/`
- `MenuUpload` record created

#### **Stage 1: LLM Menu Extraction** (`/llm/extract-menu`)
- Gemini extracts dishes from the menu
- For each dish, extracts:
  - Name, description, price
  - Category/section
  - Special notes, allergen warnings
  - Prominence score
- Creates `Recipe` records with `status='needs_review'`
- Creates `MenuSection` records
- Links recipes to sections via `MenuSectionRecipe`

#### **Stage 2: LLM Ingredient Deduction** (`/llm/deduce-ingredients`)
- For each recipe, Gemini infers likely ingredients
- For each ingredient:
  - **Creates a NEW ingredient record** with `source='llm'` and unique code `llm:<uuid>`
  - Extracts quantity, unit, notes
  - Identifies allergens (using canonical list: wheat, milk, eggs, etc.)
- Links ingredients to recipes via `RecipeIngredient`
- **IMPORTANT:** Each recipe gets its own ingredient instances (no deduplication at this stage)

### Key Code Locations
```
apps/api/app/services/menu_upload.py          # Pipeline orchestration
apps/api/app/services/gemini_client.py        # LLM API client
apps/api/app/routes.py                        # /llm/extract-menu, /llm/deduce-ingredients
```

## 3. Ingredient Duplication Analysis

### Current Behavior: NO Deduplication

**When a recipe is analyzed by the LLM:**
1. The LLM identifies ingredients (e.g., "soy sauce", "wheat flour")
2. For EACH ingredient, a **NEW** `Ingredient` record is created with:
   - `code = 'llm:<random_uuid>'` (e.g., `'llm:a1b2c3d4...'`)
   - `name = <ingredient_name>` (e.g., `'soy sauce'`)
   - `source = 'llm'`
3. This new ingredient is linked to the recipe via `RecipeIngredient`

**Result:** If 100 recipes all contain "soy sauce", there will be **100 separate ingredient records** for soy sauce in the database.

### Why This Happens
See `apps/api/app/services/menu_upload.py`, line 406:
```python
ingredient_code = f"llm:{uuid4().hex}"  # Always generates unique code
ingredient_id = await dal.insert_ingredient(
    session,
    code=ingredient_code,  # Unique every time
    name=name,
    source="llm",
)
```

### OFF Ingredients: Are They Used?

**Current Implementation:** OFF ingredients are **NOT** used to match LLM-detected ingredients.

There is NO code that:
- Looks up existing OFF ingredients by name
- Matches LLM ingredient names to OFF canonical names
- Reuses OFF ingredient IDs for recipes

**Potential Use Case:** You could modify the code to:
1. When LLM identifies an ingredient, search the `ingredient` table for `source='off'` with similar name
2. If found, reuse that `ingredient_id` instead of creating a new one
3. This would provide canonical ingredient names and reduce duplication

### Product Table: Is It Used?

**Current Implementation:** The `product` table is **NOT** used anywhere.

There is NO:
- Product lookup API endpoint used by the frontend
- Barcode scanning feature
- Product search functionality

The product data is imported but never queried in production.


````python
# Analyze ingredient duplication
print("=== Ingredient Duplication Analysis ===")
print()

duplication = query("""
    SELECT 
        LOWER(TRIM(name)) as normalized_name,
        COUNT(*) as duplicate_count,
        MIN(source) as sources
    FROM ingredient
    WHERE source = 'llm'
    GROUP BY LOWER(TRIM(name))
    HAVING COUNT(*) > 1
    ORDER BY duplicate_count DESC
    LIMIT 20
""")

if len(duplication) > 0:
    print("Top duplicated ingredient names:")
    print(duplication)
    print()
    total_dupes = duplication['duplicate_count'].sum() - len(duplication)
    print(f"Total redundant ingredient records: ~{total_dupes:,}")
else:
    print("No LLM ingredients found yet. Upload a menu to see duplication.")
````

````python
# Example: Show all instances of a common ingredient
print("=== Example: All instances of 'egg' ===")
query("""
    SELECT 
        i.id,
        i.code,
        i.name,
        i.source,
        COUNT(ri.recipe_id) as used_in_recipes
    FROM ingredient i
    LEFT JOIN recipe_ingredient ri ON ri.ingredient_id = i.id
    WHERE LOWER(i.name) LIKE '%egg%' AND i.source = 'llm'
    GROUP BY i.id, i.code, i.name, i.source
    LIMIT 20
""")
````

## 4. Database Schema & Storage Analysis


````python
# Table size analysis
print("=== Database Size Analysis ===")
print()

size_query = query("""
    SELECT 
        table_name,
        pg_size_pretty(pg_total_relation_size(quote_ident(table_name))) as size
    FROM (
        SELECT 'ingredient' as table_name
        UNION ALL SELECT 'allergen'
        UNION ALL SELECT 'product'
        UNION ALL SELECT 'recipe'
        UNION ALL SELECT 'product_ingredient'
        UNION ALL SELECT 'recipe_ingredient'
        UNION ALL SELECT 'ingredient_allergen'
        UNION ALL SELECT 'product_allergen'
    ) tables
    ORDER BY pg_total_relation_size(quote_ident(table_name)) DESC
""")

print(size_query)
print()
print("NOTE: Largest tables are typically product, product_ingredient, and ingredient (if OFF imported)")
````

````python
# Row count summary
print("=== Complete Row Count Summary ===")
print()

summary = query("""
    SELECT 'ingredient (OFF)' as table_name, COUNT(*) as rows FROM ingredient WHERE source = 'off'
    UNION ALL
    SELECT 'ingredient (LLM)', COUNT(*) FROM ingredient WHERE source = 'llm'
    UNION ALL
    SELECT 'allergen', COUNT(*) FROM allergen
    UNION ALL
    SELECT 'product', COUNT(*) FROM product
    UNION ALL
    SELECT 'recipe', COUNT(*) FROM recipe
    UNION ALL
    SELECT 'product_ingredient', COUNT(*) FROM product_ingredient
    UNION ALL
    SELECT 'recipe_ingredient', COUNT(*) FROM recipe_ingredient
    UNION ALL
    SELECT 'ingredient_allergen', COUNT(*) FROM ingredient_allergen
    UNION ALL
    SELECT 'product_allergen', COUNT(*) FROM product_allergen
    UNION ALL
    SELECT 'menu_upload', COUNT(*) FROM menu_upload
    ORDER BY rows DESC
""")

print(summary)
````

## 5. Key Findings & Recommendations

### Current State Summary

✅ **What IS being used:**
- LLM-generated ingredients (`source='llm'`)
- Recipe-ingredient links
- Allergen canonical list (shared by both OFF and LLM)
- Menu upload pipeline with Gemini

❌ **What is NOT being used:**
- Product table (3.5M rows, **never queried in production**)
- Product-ingredient links (millions of rows, **never queried**)
- Product-allergen links (never queried)
- OFF ingredients (canonical taxonomy exists but **not matched to recipes**)

### Issues Identified

1. **Ingredient Duplication:**
   - Every recipe creates its own ingredient instances
   - "Soy sauce" in 50 recipes → 50 duplicate ingredient rows
   - No canonical matching or deduplication
   - Code location: `apps/api/app/services/menu_upload.py`, line 406

2. **Large Unused Dataset:**
   - Product table is ~10GB+ after full import (3.5M products)
   - Product-ingredient links are even larger (millions of relationships)
   - This data is downloaded, processed, stored, but **never accessed**

3. **OFF Ingredients Unused:**
   - OFF ingredient taxonomy could provide canonical names
   - OFF allergen mappings could improve accuracy
   - But there's **no code** to match LLM ingredients to OFF ingredients
   - Recipes never reference OFF ingredients (confirmed by query above)

### Recommendations

#### **Option 1: Remove OFF Product Data (Keep Ingredients) ⭐ RECOMMENDED**

**What to remove:**
- ❌ `product` table
- ❌ `product_ingredient` table
- ❌ `product_allergen` table
- ❌ `product_nutrition` table

**What to keep:**
- ✅ `ingredient` table (OFF taxonomy as reference)
- ✅ `allergen` table
- ✅ `ingredient_allergen` table

**Benefits:**
- Massive storage savings (~90% reduction)
- Faster database dumps/restores
- Lower hosting costs
- Faster deploys
- No impact on current functionality

**Implementation:**
```sql
-- Drop product tables
DROP TABLE IF EXISTS product_allergen CASCADE;
DROP TABLE IF EXISTS product_ingredient CASCADE;
DROP TABLE IF EXISTS product_nutrition CASCADE;
DROP TABLE IF EXISTS product CASCADE;
```

**Modify import script:**
- Keep Steps 2 & 3 (allergen & ingredient extraction)
- Skip Step 4 (product import)

**Optional enhancement - Add ingredient matching:**
```python
# In menu_upload.py, instead of always creating new ingredient:
existing = await dal.get_ingredient_by_name(session, name, exact=False)
if existing and existing.ingredient.source == 'off':
    ingredient_id = existing.ingredient.id  # Reuse OFF ingredient
else:
    ingredient_id = await dal.insert_ingredient(
        session, code=f"llm:{uuid4().hex}", name=name, source="llm"
    )
```

#### **Option 2: Remove ALL OFF Data**

**What to remove:**
- ❌ All OFF data (products, ingredients, allergens)
- ❌ Import scripts
- ❌ `data_pipeline/` directory

**What to keep:**
- ✅ Only LLM-generated data
- ✅ Recipe system

**Benefits:**
- Maximum simplicity
- Minimal storage
- No OFF dependency
- Faster development

**Drawbacks:**

```sql

DELETE FROM ingredient_allergen WHERE source = 'off';
DELETE FROM ingredient WHERE source = 'off';

DROP TABLE IF EXISTS product_ingredient CASCADE;
DROP TABLE IF EXISTS product_nutrition CASCADE;

```

#### **Option 3: Implement Full Ingredient Deduplication**

**What to change:**
- ✅ Keep OFF ingredients as canonical reference
- ✅ Match LLM-detected ingredients to OFF taxonomy
- ✅ Reuse ingredient IDs across recipes
- ❌ Remove products (still not needed)

**Benefits:**
- Cleaner data model
- Better allergen accuracy
- Easier analytics ("all recipes with wheat")
- Reduced storage for LLM ingredients

**Drawbacks:**
- More complex implementation
- Requires fuzzy name matching
- Synonym handling needed
- Migration for existing LLM ingredients

**Implementation complexity:** Medium-High

---

### My Recommendation: **Option 1**

**Start with Option 1 for these reasons:**

1. **Quick Win:** Remove product tables → immediate 90%+ storage reduction
2. **No Risk:** Doesn't affect current functionality at all
3. **Future-Proof:** Keeps OFF ingredients as reference for later use
4. **Reversible:** Easy to re-import if needed (though unlikely)

**Next steps after Option 1:**
- Monitor storage and performance
- Optionally implement ingredient matching (use OFF taxonomy)
- Consider deduplication for LLM ingredients

**Implementation priority:**
1. Backup database
2. Drop product tables (SQL above)
3. Modify import script to skip Step 4
4. Update documentation
